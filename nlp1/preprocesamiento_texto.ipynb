# Trabajando con expresiones regulares

```python
# findall: Devuelve todas las coincidencias no superpuestas de un patron en un cadena de textos
# Vamos a buscar palabras especificas

texto="Los animales de la granja incluyen vacas, cabras y gallinas."

# Patron que permite encontrar palabras que empiecen con la letra g

patron=r'\bg\w*\b'

# Usando findall para encontrarlo en la cadena

palabras_con_g=re.findall(patron,texto)

print(palabras_con_g)
```

```python
# Intente identificar otro patron sencillo en el texto

# Patron que permite encontrar palabras que terminan con la letra s
patron = r'\b\w*s\b'

palabras_con_s = re.findall(patron, texto)

print(palabras_con_s)
```

```python

texto="Mi celular es 849-629-0308 y vivo en la calle 14 numero 17, 4ta planta. \
       El de mi esposo es 212-529-0208 y trabaja en la 10ma plata de un edificio alto."

patron=r'\d+'

numeros=re.findall(patron,texto)

print(numeros)
```

```python
patron = r'\d{3}-\d{3}-\d{4}'

celular=re.findall(patron,texto)

print(celular)
```

```python
# re.sub

texto_reemplazado = re.sub(patron, 'TELEFONO', texto)
print(texto_reemplazado)
```

```python
# Sustituyan ustedes del texto los numero ordinales que aparecen utilizando la palabra ORDINAL

patron = r'\b\d+(?:ra|ro|do|da|er|ta|to|ma)\b'

texto_sin_ordinales = re.sub(patron, 'ORDINAL', texto)

print(texto_sin_ordinales)
```

```python
# re.split: Divide cadenas de texto segun un patron

texto='¡Hola! ¿Cómo estás? Estoy muy bien. ¿Cómo estas tú?'

patron=r'[!.?]+'

oraciones=re.split(patron,texto)

print(oraciones)
```

```python
# re.search: Busca un patron en la cadena y devuelve un objeto con la coincidencia

texto="Los animales del zoologico incluyen a las zebras, zorros y monos."

# Buscamos palabras que empiecen con z

patron=r'\bz\w*\b'

coincidencia = re.search(patron,texto)

if coincidencia:
    
    print(f'Palabra encontrada:{coincidencia.group()}')
    
else:
    print('No encontre ninguna coincidencia')
```

```python
# re.match: Verifica si la cadena coincide con el patron desde el inicio de la cadena

texto ='juan.perez@gmail.com'
patron=r'^\w+([.]?\w+)*@\w+([.]?\w+)*(\.\w{2,3})+$'
correo=re.match(patron,texto)
if correo:
    print(f'Encontre correo:{correo.group()}')
    
else:
    print('No encontre correo')
```

# Distancia entre palabras (EDIT DISTANCE)

```python
# Levenstein
cadena1='gato'
cadena2='zurigata'
distancia=Levenshtein.distance(cadena1,cadena2)
print(f"El numero de operaciones para hacer iguales las cadenas es {distancia}")
```

```python
# Jaro-Winkler
cadena1='polo'
cadena2='pata'
distancia=jellyfish.jaro_similarity(cadena1,cadena2)
print(f"Distancia de Jaro-Winkler es: {distancia}")
```

```python
# Indice de Hamming
cadena1='caso'
cadena2='cama'
distancia=distance.hamming(cadena1,cadena2)
print(f"Distancia de Hamming es: {distancia}")
```
