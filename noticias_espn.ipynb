```python
import pandas as pd
import feedparser
```

```python
# Guardamos varias secciones del RSS de ESPN en un diccionario
secciones = {'general': 'https://www.espn.com/espn/rss/news',
    'nba': 'https://www.espn.com/espn/rss/nba/news',
    'nfl': 'https://www.espn.com/espn/rss/nfl/news',
    'mlb': 'https://www.espn.com/espn/rss/mlb/news'
}
lista_noticias = []

# Leemos cada feed con feedparser
for seccion, url in secciones.items():
    feed = feedparser.parse(url)
    
    for noticia in feed.entries:
        lista_noticias.append({
            'seccion': seccion,
            'titulo': noticia.title,
            'descripcion': noticia.description,
            'enlace': noticia.link,
            'fecha': noticia.published
        })

# Convertimos la lista en un dataframe de pandas
noticias_espn = pd.DataFrame(lista_noticias)

noticias_espn.head()
```

```python
noticias_espn.to_csv('espn_rss_noticias.csv', encoding='UTF-8', index=False)
```
